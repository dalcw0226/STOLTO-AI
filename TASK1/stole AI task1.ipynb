### stolio 첫 번째 과제에 대한 정답

```python
# 라이브러리 임포트
import sklearn
from sklearn.datasets import load_diabetes
import numpy as np
import matplotlib.pyplot as plt

# 데이터 로드
diabetes = load_diabetes()

data = diabetes['data']
target  = diabetes['target']
```

#### 문제 3
**인덱스 2번, 8번, 9번 feature에 대한 선형회귀 모델을 만드시오**  
데이터를 train_input, train_target, test_input, test_target으로 쪼개서 진행할 것  
sklearn을 이용!  
산점도 위에 모델이 예측한 선을 그릴것  
과적합/과소적합 여부를 확인할 것  
모델의 score점수를 알아낼 것  
coef_와 intercept_ 를 알아낼 것  

```python
# 선형 회귀를 위한 라이브러리 임포트
from sklearn.linear_model import LinearRegression

# 데이터를 쪼개기 위한 라이브러리 임포트
from sklearn.model_selection import train_test_split

# idx 2
# input 데이터와 target 데이터를 만든다.
input_idx_2 = data[:, 2].reshape(-1, 1)  # 데이터의 차원을 2차원으로 변경
target_idx_2 = target

# 데이터 쪼개기
train_input_idx_2, test_input_idx_2, train_target_idx_2, test_target_idx_2 = train_test_split(input_idx_2, target_idx_2)

# 모델 생성
lr_idx_2 = LinearRegression()
lr_idx_2.fit(train_input_idx_2, train_target_idx_2)

# 모델 평가
print(lr_idx_2.score(train_input_idx_2, train_target_idx_2))
print(lr_idx_2.score(test_input_idx_2, test_target_idx_2))
# 일단 과소적합의 경향을 보인다.
# 모델의 성능이 매우 좋지 못함
# 데이터가 많아서 어쩔 수 없는 부분인 것 같음

# 모델 파라미터 확인하기
print(lr_idx_2.coef_, lr_idx_2.intercept_)

# 그래프에 직선 그리기
# 나는 모든 데이터를 다 사용할 것이다.
plt.scatter(input_idx_2, target_idx_2)
plt.plot([min(input_idx_2), max(input_idx_2)], [min(input_idx_2)*lr_idx_2.coef_+lr_idx_2.intercept_, max(input_idx_2)*lr_idx_2.coef_+lr_idx_2.intercept_])
plt.show()
```

```python
# 그냥 예쁘게 함수형 프로그래밍을 할 수도 있다.
def answer3(idx, data, target):
    # idx 2
    # input 데이터와 target 데이터를 만든다.
    input_ = data[:, idx].reshape(-1, 1)  # 데이터의 차원을 2차원으로 변경
    target_ = target

    # 데이터 쪼개기
    train_input, test_input, train_target, test_target = train_test_split(input_, target_)

    # 모델 생성
    lr = LinearRegression()
    lr.fit(train_input, train_target)

    # 모델 평가
    print(lr.score(train_input, train_target))
    print(lr.score(test_input, test_target))

    # 모델 파라미터 확인하기
    print(lr.coef_, lr.intercept_)

    # 그래프에 직선 그리기
    # 나는 모든 데이터를 다 사용할 것이다.
    plt.scatter(input_, target_)
    plt.plot([min(input_), max(input_)], [min(input_)*lr.coef_+lr.intercept_, max(input_)*lr.coef_+lr.intercept_])
    plt.show()
    
    print("\n") # 구분을 위한 행
    
# 정답 출력
for idx in [2, 8, 9]:
    answer3(idx, data, target)
```

#### 문제 4
**인덱스 6번 featur에 대한 다항회귀모델을 만드시오(dim = 2)**  
데이터를 train_input, train_target, test_input, test_target으로 쪼개서 진행할 것  
sklearn을 이용!  
산점도 위에 모델이 예측한 선을 그릴것  
과적합/과소적합여부를 판단할 것  
모델의 score점수를 알아낼 것  
coef_와 intercept_ 를 알아낼 것  

```python
# 선형 회귀를 위한 라이브러리 임포트
from sklearn.linear_model import LinearRegression

# 데이터를 쪼개기 위한 라이브러리 임포트
from sklearn.model_selection import train_test_split

input_ = np.column_stack((data[:,6]**2, data[:, 6]))
target_ = target

# 데이터 쪼개기
train_input, test_input, train_target, test_target = train_test_split(input_, target_)

# 모델 생성
lr = LinearRegression()
lr.fit(train_input, train_target)

# 모델 파라미터 확인
print(lr.coef_, lr.intercept_)

# 성능 평가
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))

# 그래프 그리기
# print(min(data[:,6]))
# point = np.arange(min(data[:,6]), max(data[:,6]))
# plt.scatter(data[:,6], target_)

# plt.plot(point, 1758*point**2-760*point+147)
# plt.show()
```

#### 문제 5
**인덱스 2번, 8번, 9번 feature을 모두 이용하여 1개의 다중 회귀 모델을 만드시오**  
데이터를 train_input, train_target, test_input, test_target으로 쪼개서 진행할 것  
sklearn을 이용!  
과적합/과소적합여부를 판단할 것  
모델의 score점수를 알아낼 것  
coef_와 intercept_ 를 알아낼 것  

```python
# 데이터
input_ = np.column_stack((data[:, 2], data[:, 8], data[:, 9]))
target_ = target

# 쪼개기
train_input, test_input, train_target, test_target = train_test_split(input_, target_)

# 모델 생성
lr = LinearRegression()
lr.fit(train_input, train_target)

# 모델 파라미터 확인
print(lr.coef_, lr.intercept_)

# 성능 평가
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```
